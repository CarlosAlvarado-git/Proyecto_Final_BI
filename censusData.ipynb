```python
import pandas as pd
import numpy as np
import os
import re

path = 'C:/GitHub/Proyecto_Final_BI/'
```

```python
# función dada en clase
def load_data(filename):
    """ Load dataset from filename
    Args:
        filename (string): Path to data
        
    Returns:
        df (DataFrame): Dataset from file
    """
    df = pd.read_csv(filename)
    return df
```

```python
def trim_all_columns(df):
    """
    Trim whitespace from ends of each value across all series in dataframe
    """
    trim_strings = lambda x: x.strip() if isinstance(x, str) else x
    return df.applymap(trim_strings)
```

```python
dfAges = pd.DataFrame()
baseLimit  = []
upperLimit = []
for item in ages:
    numbers = re.findall(r'\d+', item)
    baseLimit.append(int(numbers[0]))
    if len(numbers) > 1:
        upperLimit.append(int(numbers[1]))
    else:
        upperLimit.append(-1)
dfAges['baseLimit'] = baseLimit
dfAges['upperLimit'] = upperLimit
dfAges['income'] = incomes
dfAges
```
